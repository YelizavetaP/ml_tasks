### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp

x = sp.symbols('x')
f = 10 * x - 5
df = sp.diff(f, x)
```

```python
def show_f(f, df):
    print(f'Похідна функції {f}:\n{df}')
```

```python
show_f(f, df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
f = 10 * x - 5
df = np.full(11, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, f, 'r', label='f(x) = 10x - 5')
plt.plot(x, df, 'b', label='df(x) = 10')

plt.show()
```

З графіку видно, що зі збільшенням кількості витрачених на навчвння годин росте і оцінка.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.log(x)

df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)

show_f(f1, df1)
show_f(f2, df2)
```

```python
x_val = np.linspace(0, 10, 100)

f1_val = np.sin(x_val)
df1_val = np.cos(x_val)

plt.plot(x_val, f1_val, 'r', label=f1)
plt.plot(x_val, df1_val, '--', color='red', label=df1)

plt.title(f"Графік функції {f1} та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
f2_val = np.log(x_val)
df2_val = 1/x_val

plt.plot(x_val, f2_val, 'b', label=f2)
plt.plot(x_val, df2_val, '--', color='b', label=df2)

plt.title(f"Графік функції {f2} та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5 * x + 1
df = sp.diff(f, x)

show_f(f, df)
```

```python
x_val = np.arange(0, 101)


f_val = x_val**2 - 5 * x_val + 1
df_val = 2 * x_val - 5

plt.plot(x_val, f_val, 'r', label=f)
plt.plot(x_val, df_val, '--', color='b', label=df)

plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Загальні витрати компанії")
plt.legend()

plt.show()
```

```python
x_val = np.linspace(0, 10, 100)

f_val = x_val**2 - 5 * x_val + 1
df_val = 2 * x_val - 5

plt.plot(x_val, f_val, 'r', label=f)
plt.plot(x_val, df_val, '--', color='b', label=df)

plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Загальні витрати компанії")
plt.legend()

plt.show()
```
