### ансамблі

Додаткові advanced методи оптимізації гіперпараметрів
Просунутий метод оптимізації гіперпараметрів - це використовуючи Баєсівську оптимізацію

* Пояснення як працює цей метод: https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74/

* Приклад використання: https://scikit-optimize.github.io/stable/auto_examples/sklearn-gridsearchcv-replacement.html

Бібліотеки для оптимізації гіперпараметрів: що використовують в індустрії крім базового sklearn

* Scikit-optimize - використовує алгоритм оптимізації на основі послідовної моделі, щоб знаходити оптимальні рішення для проблем пошуку гіперпараметрів за менший час.

Scikit-optimize надає багато функцій, крім оптимізації гіперпараметрів, таких як збереження та завантаження результатів оптимізації, графіки збіжності моделей, порівняння сурогатних моделей (для Bayesian Optimization)

документація: https://scikit-optimize.github.io/stable/index.html

* Optuna - використовує історичний запис деталей слідів, щоб визначити перспективну область для пошуку оптимізації гіперпараметра і, отже, знаходить оптимальний гіперпараметр за мінімальний проміжок часу.

Optuna має функцію обрізки (prunning), яка автоматично зупиняє безперспективні стежки на ранніх етапах навчання:

документація: https://optuna.readthedocs.io/en/stable/

приклад тюнингу параметрів для xgboost: https://forecastegy.com/posts/xgboost-hyperparameter-tuning-with-optuna/

* Hyperopt - бібліотека для байєсової оптимізації гіперпараметрів. 

документація: https://hyperopt.github.io/hyperopt/

приклад тюнингу буде в Демо роботи xgboost та lightGBM
